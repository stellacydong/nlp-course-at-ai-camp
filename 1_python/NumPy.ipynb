From the example above, you can see that lists can only be combined. Lists cannot be added in the sense of our normal math operation. Therefore we need numpy.

### application of numpy arrays in computer vision 

```python
# 0 black, 255 white, 0-255 gray 
zero_array = np.zeros((3,5))  # 3 rows, 5 columns 
```

```python
print(zero_array)
```

```python
# import matplotlib to plot 
import matplotlib.pyplot as plt 
```

```python
plt.imshow(zero_array, cmap = 'gray')  # cmap = color map = gray 
```

```python
# white image (3 dimensional array)
white_img = np.full((5,5,3), 255)  # 5 rows, 5 columns, 3 channels = Red, Green, Blue (RGB)
```

```python
print(white_img)
```

```python
white_img[:,:,0]  # the first channel, red channel, it shall be a 5x5 array, all entries = 255
```

```python
white_img[:,:,1] # 2nd channel, green channel, 5x5 array 
```

```python
white_img[:,:,2] # 3rd channel, blue channel, 5x5
```

```python
plt.imshow(np.uint8(white_img), cmap = 'gray')
plt.grid('on')
```
